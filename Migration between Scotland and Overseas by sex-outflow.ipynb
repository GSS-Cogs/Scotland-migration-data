```python
from databaker.framework import *
import pandas as pd 
```

```python
import requests
from cachecontrol import CacheControl
from cachecontrol.caches.file_cache import FileCache
from cachecontrol.heuristics import LastModified
from pathlib import Path

session = CacheControl(requests.Session(),
                       cache=FileCache('.cache'),
                       heuristic=LastModified())

sourceFolder = Path('in')
sourceFolder.mkdir(exist_ok=True)

inputURL = 'https://www.nrscotland.gov.uk/files//statistics/migration/2018-july/tab-z1-overseas-mig-flows-admin-sex-hb-2001-02-latest-july-18.xlsx'
inputFile = sourceFolder / 'tab-z1-overseas-mig-flows-admin-sex-hb-2001-02-latest-july-18.xlsx'
response = session.get(inputURL)
with open(inputFile, 'wb') as f:
  f.write(response.content)
```

```python
tab = loadxlstabs(inputFile, sheetids='Out-Council Area-Sex')[0]
```

```python
observations = tab.excel_ref("B6").expand(RIGHT).expand(DOWN).is_not_blank().is_not_whitespace()
observations 
```

```python
observations = observations - tab.excel_ref('B').expand(DOWN).by_index([40,41,42,77,78,79])
```

```python
area = tab.excel_ref('A6').expand(DOWN).is_not_blank() - tab.excel_ref('A').expand(DOWN).by_index([40,41,42,77,78,79])
area
```

```python
flow = tab.excel_ref('B').expand(DOWN).by_index([3,40,77])
flow
```

```python
midyear = tab.excel_ref('B5').expand(RIGHT).is_not_blank()
midyear
```

```python
Dimensions = [
            HDim(midyear,'Mid Year',DIRECTLY,ABOVE),
            HDim(area,'Area of Destination or Origin', DIRECTLY, LEFT),
            HDim(flow,'flow',CLOSEST,ABOVE),
            HDimConst('Measure Type', 'Count'),
            HDimConst('Unit','People'),
            HDimConst('Flow','Outflow'),
            HDimConst('Age', 'all')
            ]
```

```python
c1 = ConversionSegment(observations, Dimensions, processTIMEUNIT=True)
# savepreviewhtml(c1)
```

```python
new_table = c1.topandas()
new_table
```

```python
new_table = new_table[new_table['Area of Destination or Origin'].isnull() == False]
```

```python
new_table = new_table[new_table['Mid Year'].isnull() == False]
```

```python
new_table.dtypes
```

```python
new_table.head()
```

```python
new_table['Mid Year'] = new_table['Mid Year'].map(lambda x: str(x)[0:4])
```

```python
new_table['Mid Year'] = new_table['Mid Year'] + '-06-30T00:00:00/P1Y'
```

```python
new_table['Sex'] = new_table['flow'].map(lambda x: str(x)[16:])
```

```python
new_table['Sex'].unique()
```

```python
new_table['Value'] = new_table['OBS'].astype(int)
```

```python
new_table['Sex'] = new_table['Sex'].map(
    lambda x: {
        'Persons' : 'T', 
        'Females' : 'F',
        'Males': 'M' 
        }.get(x, x))

```
